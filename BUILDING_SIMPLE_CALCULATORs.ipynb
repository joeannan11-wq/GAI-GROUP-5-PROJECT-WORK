```python
#FUNCTIONS INTRODUCTION
- Inputs
- Processing part
- Output

to create a function in python use the def keyword
```

```python
import gradio as gr
def add_numbers(num_1,num_2):
  return num_1 + num_2


app = gr.Interface(
    fn = add_numbers,
    inputs = [gr.Number(label="Number 1"), gr.Number(label="Number 2")],
    outputs = gr.Number(label="Result"),
    title = "Simple calculator"

)
app.launch()
```
