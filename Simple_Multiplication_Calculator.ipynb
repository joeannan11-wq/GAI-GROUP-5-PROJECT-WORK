```python
# Simple Multiplication Calculator
```

```python
import gradio as gr
def multiply_numbers(num_1,num_2):
  return num_1 * num_2


app = gr.Interface(
    fn = multiply_numbers,
    inputs = [gr.Number(label="Number 1"), gr.Number(label="Number 2")],
    outputs = gr.Number(label="Result"),
    title = "Simple calculator"

)
app.launch()
```
